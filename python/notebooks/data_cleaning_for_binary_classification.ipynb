```python
import pandas as pd
import numpy as np
```

```python
df_crashes = pd.read_csv(r'..\..\data\motor_vehicle_collisions_crashes_with_precinct.csv', parse_dates=[['CRASH DATE', 'CRASH TIME']])
#df_crashes  pd.read_csv(r'..\..\data\Motor_Vehicle_Collisions_-_Crashes.csv')
```

```python
df_crashes.columns
```

```python
#df_crashes['crash_year'] = pd.DatetimeIndex(df_crashes['CRASH DATE_CRASH TIME']).year
#df_crashes['crash_month'] = pd.DatetimeIndex(df_crashes['CRASH DATE_CRASH TIME']).month
#df_crashes['crash_month_year'] = pd.to_datetime(df_crashes['CRASH DATE_CRASH TIME']).dt.to_period('M')
df_crashes['crash_month_year_hour'] = pd.to_datetime(df_crashes['CRASH DATE_CRASH TIME']).dt.to_period('H')
#df_crashes['crash_day'] = pd.DatetimeIndex(df_crashes['CRASH DATE_CRASH TIME']).day
#df_crashes['crash_hour'] = pd.DatetimeIndex(df_crashes['CRASH DATE_CRASH TIME']).hour
#df_crashes['datetime'] = pd.DatetimeIndex(df_crashes['CRASH DATE'])
```

```python
df_crashes
```

```python
#remove dates which are 2012 or 2020 
df_crashes = df_crashes.loc[(df_crashes['CRASH DATE_CRASH TIME'] > '2013-01-01') & (df_crashes['CRASH DATE_CRASH TIME'] < '2020-01-01') ]
```

```python
#print(df_crashes['crash_month_year'].min())
#print(df_crashes['crash_month_year'].max())
```

```python
#drop crashes that are earlier than 2013-01 and later than 2019-12
#df_crashes = df_crashes.loc[(df_crashes['crash_month_year'] > '2013-01') & (df_crashes['crash_month_year'] < '2020-01')]
```

```python
df_precinct_not_null = df_crashes.loc[
    (df_crashes['precinct'].notnull()) & (df_crashes['precinct'].notna())
]
```

```python
df_features = df_precinct_not_null[[
    'CRASH DATE_CRASH TIME',
    'crash_month_year_hour',
    'precinct',
    'crash_year',
    'crash_month',
    'crash_day',
    'crash_hour'
]]
```

```python
#add a column called 'crash_binary'
df_features['crash_binary'] = 1
```

```python
#to prepare this dataset for binary classification, we will generate a response column (1 or 0) 
#1 denotes crash
# 0 denotes no crash

#for every year, every month, every day, and every hour (per borough) there will be an entry
#aggregate 
df_features
```

```python
#df_sum_of_crashes = df_features_manhattan.groupby(['crash_year', 'crash_month', 'crash_day', 'crash_hour'], axis=0)\
#    .agg({'crash_binary': ['sum']})
df_sum_of_crashes = df_features.groupby(['precinct','crash_month_year_hour'], axis=0)\
    .agg({'crash_binary': ['sum']})
df_sum_of_crashes
```

```python
#df_sum_of_crashes.reset_index(inplace=True)
#print(example)
#type(example)
#df_sum_of_crashes = df_sum_of_crashes.sort_values(by=['crash_year', 'crash_month', 'crash_day', 'crash_hour'])
df_sum_of_crashes = df_sum_of_crashes.sort_values(by=['precinct','crash_month_year_hour'])
df_sum_of_crashes
```

```python
test_df = df_sum_of_crashes.copy()
print(df_sum_of_crashes.head())
print(test_df.head())
```

```python
#test_df.index = test_df[['precinct', 'crash_month_year_hour']]
#test_df
test_df.index
```

```python
test_df.columns
```

```python
datetime_hour_index = pd.date_range(start='1/1/2013', end='12/31/2019 23:00', freq = 'H').to_period('H')
mux = pd.MultiIndex.from_product([test_df.index.levels[0], datetime_hour_index], 
                                 names=test_df.index.names)
test_df_reindexed  = test_df.reindex(mux)
test_df_reindexed
```

```python
combined_df = df_sum_of_crashes.join(test_df_reindexed, how='outer', lsuffix='_left', rsuffix='_right')
```

```python
combined_df.columns
```

```python
combined_df[['crash_binary_left']].isnull().sum()
```

```python
combined_df[['crash_binary_right']].isnull().sum()
```

```python
len(combined_df)
```

```python
(combined_df[['crash_binary_right']] == 1).sum()
```

```python
#Ereset the index, we don't want multiindex anymore
combined_df_reset_index = combined_df.reset_index()
```

```python
#fill NaN with 0's
combined_df_fill_na = combined_df_reset_index.fillna(0)
combined_df_fill_na
```

```python
combined_df_fill_na['timestamp'] = combined_df_fill_na['crash_month_year_hour'].apply(lambda x: x.to_timestamp())
```

```python
#combined_df_fill_na['timestamp_est'] = combined_df_fill_na['timestamp'].apply(lambda x: x.tz_localize(tz='EST'))
```

```python
combined_df_fill_na
```

```python
combined_df_fill_na['month'] = pd.DatetimeIndex(combined_df_fill_na['timestamp'], tz='EST').month
combined_df_fill_na['week'] = pd.DatetimeIndex(combined_df_fill_na['timestamp'], tz='EST').week
combined_df_fill_na['day'] = pd.DatetimeIndex(combined_df_fill_na['timestamp'], tz='EST').day
combined_df_fill_na['weekday'] = pd.DatetimeIndex(combined_df_fill_na['timestamp'], tz='EST').weekday
combined_df_fill_na['hour'] = pd.DatetimeIndex(combined_df_fill_na['timestamp'], tz='EST').hour
```

```python
combined_df_fill_na
```

```python
combined_df_fill_na['total_number_of_crashes'] = combined_df_fill_na[('crash_binary_right', 'sum')]
```

```python
combined_df_fill_na['did_crash_happen'] = combined_df_fill_na['total_number_of_crashes'].apply(lambda x: 1 if (x > 0) else 0 )
```

```python
#sanity check
(combined_df_fill_na['did_crash_happen'] == 0).sum()
```

```python
#sanity check #2
(combined_df_fill_na['total_number_of_crashes'] == 0).sum()
```

```python
final_df = combined_df_fill_na[['precinct', 'timestamp', 'month', 'week', 'day', 'weekday', 'hour', 'total_number_of_crashes', 'did_crash_happen']]
final_df
```

```python
final_df.to_csv('..\..\data\motor_vehicle_collisions_crashes_cleaned_v2.csv', \
                header=['precinct', 'timestamp', 'month', 'week', 'day', 'weekday', 'hour', 'total_number_of_crashes', 'did_crash_happen'],\
               index_label='id')
```
